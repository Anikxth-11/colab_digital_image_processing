```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('/content/WhatsApp Image 2024-09-04 at 8.21.50 PM.jpeg', cv2.IMREAD_GRAYSCALE)
image_color = cv2.imread('/content/WhatsApp Image 2024-09-04 at 8.21.50 PM.jpeg')
```

```python
def display_image(title, img):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def negative_image(image):
    negative = 255 - image
    return negative
```

```python
def flip_image(image, direction):
    flipped_image = cv2.flip(image, direction)
    return flipped_image
```

```python
def threshold_image(image, thresh_value):
    _, thresholded = cv2.threshold(image, thresh_value, 255, cv2.THRESH_BINARY)
    return thresholded
```

```python
def contrast_stretch(image):
    min_val = np.min(image)
    max_val = np.max(image)
    contrast_stretched = ((image - min_val) / (max_val - min_val)) * 255
    contrast_stretched = np.uint8(contrast_stretched)
    return contrast_stretched
```

```python
display_image('Original Image', image)
```

```python
negative_img = negative_image(image)
display_image('Negative Image', negative_img)
```

```python
flipped_img_h = flip_image(image, 1)
display_image('Horizontally Flipped Image', flipped_img_h)
```

```python
flipped_img_v = flip_image(image, 0)
display_image('Vertically Flipped Image', flipped_img_v)
```

```python
thresholded_img = threshold_image(image, 127)
display_image('Thresholded Image (Threshold = 127)', thresholded_img)
```

```python
contrast_stretched_img = contrast_stretch(image)
display_image('Contrast Stretched Image', contrast_stretched_img)
```
