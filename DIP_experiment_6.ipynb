```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
original_image = cv2.imread('/content/WA.jpeg', cv2.IMREAD_GRAYSCALE)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
equalized_image = clahe.apply(original_image)
```

```python
def generate_low_contrast_image(size=(256, 256)):
    """Generates a low contrast grayscale image."""
    # Create an image with pixel values between 50 and 100
    return np.random.randint(50, 101, size, dtype=np.uint8)
```

```python
def generate_high_contrast_image(size=(256, 256)):
    """Generates a high contrast grayscale image."""
    # Create an image with pixel values spread across the full range
    return np.random.randint(0, 256, size, dtype=np.uint8)
```

```python
def generate_dark_image(size=(256, 256)):
    """Generates a dark grayscale image."""
    # Create an image with pixel values between 0 and 50
    return np.random.randint(0, 51, size, dtype=np.uint8)
```

```python
def generate_bright_image(size=(256, 256)):
    """Generates a bright grayscale image."""
    # Create an image with pixel values between 200 and 255
    return np.random.randint(200, 256, size, dtype=np.uint8)
```

```python
def plot_histogram(image, title, ax):
    """Plots the histogram of a grayscale image."""
    ax.hist(image.flatten(), bins=256, range=[0,256], color='gray')
    ax.set_title(title)
    ax.set_xlim([0,256])
```

```python
def apply_histogram_equalization(image):
    """Applies histogram equalization to a grayscale image."""
    return cv2.equalizeHist(image)
```

```python
def display_results(original, equalized, title, idx):
    """Displays the original and equalized images along with their histograms."""
    plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.imshow(original_image, cmap='gray', vmin=0, vmax=255)
plt.title(f'Original {original_image} Image')
plt.axis('off')
```

```python
plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray', vmin=0, vmax=255)
plt.title(f'Equalized {equalized_image} Image')
plt.axis('off')
```

```python
plt.subplot(2, 2, 3)
plot_histogram(original_image, f'Original {original_image} Histogram', plt.gca())
```

```python
plt.subplot(2, 2, 4)
plot_histogram(equalized_image, f'Equalized {equalized_image} Histogram', plt.gca())

plt.tight_layout()
plt.show()
```

```python
def main():
    # Define image types and corresponding generation functions
    image_types = {
        'Low Contrast': generate_low_contrast_image,
        'High Contrast': generate_high_contrast_image,
        'Dark': generate_dark_image,
        'Bright': generate_bright_image
    }

    for idx, (title, func) in enumerate(image_types.items(), 1):
        # Generate the original image
        original_image = func()

        # Apply histogram equalization
        equalized_image = apply_histogram_equalization(original_image)

        # Display the results
        display_results(original_image, equalized_image, title, idx)
```

```python
if __name__ == "__main__":
    main()
```
