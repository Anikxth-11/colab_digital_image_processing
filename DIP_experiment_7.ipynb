```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image (assumed to be noisy)
image = cv2.imread('/content/123.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
kernel_sizes = [3, 5, 9]
filtered_images = []

for k in kernel_sizes:
    filtered_image = cv2.blur(image, (k, k))  # Averaging filter
    filtered_images.append(filtered_image)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
```

```python
for i, k in enumerate(kernel_sizes):
    plt.subplot(1, 4, i + 2)
    plt.imshow(filtered_images[i], cmap='gray')
    plt.title(f'Averaging {k}x{k}')

plt.show()
```

```python
# Apply median filter with different kernel sizes
filtered_images_median = []

for k in kernel_sizes:
    filtered_image_median = cv2.medianBlur(image, k)  # Median filter
    filtered_images_median.append(filtered_image_median)
```

```python
# Display the original and median filtered images
plt.figure(figsize=(10, 5))
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
```

```python
for i, k in enumerate(kernel_sizes):
    plt.subplot(1, 4, i + 2)
    plt.imshow(filtered_images_median[i], cmap='gray')
    plt.title(f'Median {k}x{k}')

plt.show()
```

```python
# Define the 3x3 high pass filter kernel for sharpening
sharpening_kernel = np.array([[0, -1, 0],
                              [-1, 5, -1],
                              [0, -1, 0]])
```

```python
# Apply the sharpening filter
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)
```

```python
# Display the original and sharpened images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(sharpened_image, cmap='gray')
plt.title("Sharpened Image")

plt.show()
```
